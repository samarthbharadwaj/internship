```python
import requests
from urllib.parse import urljoin
import re

class VulnerabilityScanner:
    def __init__(self, target_url):
        self.target = target_url
        self.session = requests.Session()
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Security Scanner)'
        }

    def check_sql_injection(self, param):
        """Test for SQL injection vulnerability"""
        payloads = [
            "' OR '1'='1",
            "1' UNION SELECT NULL--",
            "admin'--",
        ]

        for payload in payloads:
            try:
                response = self.session.get(
                    self.target,
                    params={param: payload},
                    headers=self.headers,
                    timeout=5
                )

                if any(x in response.text.lower() for x in ['syntax', 'error', 'exception']):
                    print(f"[!] Potential SQL injection on {param}")
                    return True
            except Exception as e:
                print(f"[-] Error testing {param}: {e}")

        return False

    def check_xss(self):
        """Test for XSS vulnerabilities"""
        xss_payload = "<script>alert('XSS')</script>"

        try:
            response = self.session.get(self.target, headers=self.headers)

            if xss_payload in response.text:
                print("[!] Potential reflected XSS vulnerability detected")
                return True
        except Exception as e:
            print(f"[-] XSS check error: {e}")

        return False

    def check_security_headers(self):
        """Check for missing security headers"""
        important_headers = [
            'Strict-Transport-Security',
            'X-Content-Type-Options',
            'X-Frame-Options',
            'Content-Security-Policy',
            'X-XSS-Protection'
        ]

        try:
            response = self.session.head(self.target, headers=self.headers, timeout=5)

            print("[*] Security Headers Check:")
            for header in important_headers:
                if header in response.headers:
                    print(f"    [+] {header}: {response.headers[header]}")
                else:
                    print(f"    [-] Missing: {header}")
        except Exception as e:
            print(f"[-] Header check error: {e}")
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer import AerSimulator
import numpy as np

class QuantumAlgorithms:
    """Quantum computing demonstrations"""

    def deutsch_algorithm(self):
        """
        Deutsch algorithm - determine if function is balanced or constant
        Demonstrates quantum advantage
        """
        print("\n[*] Deutsch Algorithm (Constant vs Balanced Function)")

        # Create quantum circuit
        qc = QuantumCircuit(2, 1, name='Deutsch')

        # Initialize
        qc.x(1)  # Set second qubit to |1вџ©
        qc.h([0, 1])  # Hadamard on both qubits

        # Oracle for balanced function (CNOT)
        qc.cx(0, 1)

        # Final Hadamard
        qc.h(0)
        qc.measure(0, 0)

        # Simulate
        simulator = AerSimulator()
        result = simulator.run(qc, shots=1000).result()
        counts = result.get_counts()

        print(f"[+] Results: {counts}")
        if '0' in counts and counts['0'] > 900:
            print("[+] Function is BALANCED")
        else:
            print("[+] Function is CONSTANT")

    def grover_search(self, marked_state=3, num_qubits=2):
        """
        Grover's algorithm - quantum search
        Search for a marked state in unsorted database
        """
        print(f"\n[*] Grover's Algorithm (searching for state {marked_state})")

        qc = QuantumCircuit(num_qubits, num_qubits)

        # Initialize superposition
        qc.h(range(num_qubits))

        # Oracle: mark the target state
        target_binary = format(marked_state, f'0{num_qubits}b')
        for i, bit in enumerate(target_binary):
            if bit == '0':
                qc.x(i)

        # Multi-controlled Z gate
        if num_qubits == 2:
            qc.cz(0, 1)

        for i, bit in enumerate(target_binary):
            if bit == '0':
                qc.x(i)

        # Diffusion operator
        qc.h(range(num_qubits))
        qc.x(range(num_qubits))
        if num_qubits == 2:
            qc.cz(0, 1)
        qc.x(range(num_qubits))
        qc.h(range(num_qubits))

        # Measure
        qc.measure(range(num_qubits), range(num_qubits))

        # Simulate
        simulator = AerSimulator()
        result = simulator.run(qc, shots=1000).result()
        counts = result.get_counts()

        print(f"[+] Results: {counts}")
        top_result = max(counts, key=counts.get)
        print(f"[+] Most likely state: {top_result} (decimal: {int(top_result, 2)})")

    def shor_factoring_simulation(self):
        """
        Simplified Shor's algorithm demonstration
        Shows how quantum computers could break RSA
        """
        print("\n[*] Shor's Algorithm Simulation")
        print("[*] (Simplified - full implementation is complex)")

        # For educational purposes, show the period-finding concept
        N = 15  # Number to factor (3 Г— 5)
        a = 7   # Random number coprime to N

        print(f"[+] Factoring N = {N} using a = {a}")

        # Classical period finding (what Shor's does quantumly)
        def find_period(a, N):
            for r in range(1, N):
                if pow(a, r, N) == 1:
                    return r
            return None

        period = find_period(a, N)
        print(f"[+] Period r = {period}")

        if period and period % 2 == 0:
            x = pow(a, period // 2, N)
            factor1 = np.gcd(x - 1, N)
            factor2 = np.gcd(x + 1, N)

            if factor1 > 1 and factor2 > 1:
                print(f"[+] Factors found: {factor1} Г— {factor2} = {factor1 * factor2}")
            else:
                print(f"[+] Potential factors: {factor1}, {factor2}")

    def quantum_key_distribution_bb84(self):
        """BB84 Quantum Key Distribution protocol simulation"""
        print("\n[*] BB84 Quantum Key Distribution")

        np.random.seed(42)
        n_qubits = 20

        # Alice generates random bits and bases
        alice_bits = np.random.randint(0, 2, n_qubits)
        alice_bases = np.random.randint(0, 2, n_qubits)  # 0=rectilinear, 1=diagonal

        # Bob randomly chooses bases
        bob_bases = np.random.randint(0, 2, n_qubits)

        # Simulate measurement (Bob gets correct result if bases match)
        bob_bits = np.zeros(n_qubits, dtype=int)
        for i in range(n_qubits):
            if alice_bases[i] == bob_bases[i]:
                bob_bits[i] = alice_bits[i]
            else:
                bob_bits[i] = np.random.randint(0, 2)

        # Sift keys (keep only matching bases)
        matching_indices = alice_bases == bob_bases
        sifted_key = alice_bits[matching_indices]

        print(f"[+] Initial qubits sent: {n_qubits}")
        print(f"[+] Matching bases: {np.sum(matching_indices)}")
        print(f"[+] Sifted key length: {len(sifted_key)}")
        print(f"[+] Sifted key: {''.join(map(str, sifted_key[:20]))}")
        print(f"[+] вњ“ Shared secret established (eavesdropping would be detected)")

# Usage
if __name__ == "__main__":
    qa = QuantumAlgorithms()

    qa.deutsch_algorithm()
    qa.grover_search()
    qa.shor_factoring_simulation()
    qa.quantum_key_distribution_bb84()
```
