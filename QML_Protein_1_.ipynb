```python
import numpy as np
import pennylane as qml
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
torsion_angles = np.random.uniform(-np.pi, np.pi, (200, 3))
targets = np.random.uniform(-1, 1, 200)
n_wires = 3
dev = qml.device("default.qubit", wires=n_wires)
@qml.qnode(dev)
def circuit(angles):
    for i in range(n_wires):
        qml.RX(angles[i], wires=i)
        qml.RY(angles[i], wires=i)
    qml.CNOT(wires=[0,1])
    qml.CNOT(wires=[1,2])
    return [qml.expval(qml.PauliZ(i)) for i in range(n_wires)]
quantum_features = np.array([circuit(angles) for angles in torsion_angles])
X_train, X_test, y_train, y_test = train_test_split(quantum_features, targets, test_size=0.2, random_state=42)
model = Ridge(alpha=0.5).fit(X_train, y_train)
print("R^2 score on test set:", model.score(X_test, y_test))
plt.figure(figsize=(10,5))
plt.plot(targets[:50], label="Target (Stability/Energy)", color="blue")
plt.plot([val[0] for val in quantum_features[:50]], label="Quantum Encoded Feature", color="orange")
plt.title("Protein Folding QML: Targets vs Quantum Encoded Features")
plt.xlabel("Samples")
plt.legend()
plt.show()
```

```python

#Quantum grapgh neural network
import pennylane as qml
from pennylane import numpy as np
import torch
import networkx as nx
protein_graph = nx.Graph()
protein_graph.add_nodes_from([0,1,2,3])  # 4 amino acids
protein_graph.add_edges_from([(0,1),(1,2),(2,3)])  # simple chain
features = {
    0: [0.2, 0.8],
    1: [0.5, 0.3],
    2: [0.9, 0.1],
    3: [0.4, 0.6]
}
n_qubits = 2
dev = qml.device("default.qubit", wires=n_qubits)
def encode_node(x):
    qml.AngleEmbedding(x, wires=range(n_qubits))
def variational_block(weights):
    qml.StronglyEntanglingLayers(weights, wires=range(n_qubits))
@qml.qnode(dev, interface="torch")
def qnode(x, weights):
    encode_node(x)
    variational_block(weights)
    return qml.expval(qml.PauliZ(0))
weights = torch.randn((2, n_qubits, 3), requires_grad=True)
def quantum_graph_forward(graph, features, weights):
    outputs = []
    for node in graph.nodes():
        x = torch.tensor(features[node], dtype=torch.float32)
        outputs.append(qnode(x, weights))
    return torch.mean(torch.stack(outputs))
label = torch.tensor(1.0)
opt = torch.optim.Adam([weights], lr=0.1)
for epoch in range(20):
    pred = quantum_graph_forward(protein_graph, features, weights)
    loss = (pred - label)**2
    opt.zero_grad()
    loss.backward()
    opt.step()
    print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}")
```

```python
#Quantum kernel similarity
import pennylane as qml
from pennylane import numpy as np
dev = qml.device("default.qubit", wires=4)
def feature_map(x):
    qml.AngleEmbedding(x, wires=range(4))
@qml.qnode(dev)
def kernel_circuit(x1, x2):
    feature_map(x1)
    qml.adjoint(feature_map)(x2)
    return qml.probs(wires=range(4))
def quantum_kernel(x1, x2):
    return kernel_circuit(x1, x2)[0]
protein_A = np.array([0.2, 0.5, 0.7, 0.1])
protein_B = np.array([0.3, 0.6, 0.2, 0.8])

print("Quantum kernel similarity:", quantum_kernel(protein_A, protein_B))
```

```python
import networkx as nx
G = nx.Graph()
G.add_edges_from([(0,1),(1,2),(2,3)])
dev = qml.device("default.qubit", wires=4)
@qml.qnode(dev)
def qgnn_layer(features, weights):
    qml.AngleEmbedding(features, wires=range(4))
    for (u,v) in G.edges():
        qml.CNOT(wires=[u,v])
    qml.StronglyEntanglingLayers(weights, wires=range(4))
    return [qml.expval(qml.PauliZ(i)) for i in range(4)]

weights = np.random.randn(2,4,3)
features = np.array([0.2,0.5,0.7,0.1])
print("QGNN output:", qgnn_layer(features, weights))
```

```python
import pennylane as qml
from pennylane import numpy as np
n_qubits = 16
dev = qml.device("default.qubit", wires=n_qubits)
def encode_input(seq_length):
    angles = np.linspace(0, np.pi, n_qubits)
    return angles[:seq_length]
@qml.qnode(dev)
def protein_circuit(params):
    # Encoding
    for i, angle in enumerate(params):
        qml.RX(angle, wires=i)
    # Entanglement
    for i in range(n_qubits-1):
        qml.CNOT(wires=[i, i+1])
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]
def protein_synthesis(seq_length):
    params = encode_input(seq_length)
    result = protein_circuit(params)
    return {
        "Input Sequence Length": seq_length,
        "Quantum Encoded State": params,
        "Protein Folding Signature": result
    }
user_input = int(input("Enter protein sequence length (1-4): "))
output = protein_synthesis(user_input)
print(output)
```

```python
import pennylane as qml
from pennylane import numpy as np
import itertools
n_qubits = 6
dev = qml.device("default.qubit", wires=n_qubits)
aa_properties = {
    "A": 0.1, "C": 0.2, "D": 0.3, "E": 0.4, "F": 0.5,
    "G": 0.6, "H": 0.7, "I": 0.8, "K": 0.9, "L": 1.0
}

def encode_sequence(seq):
    # Map amino acids to rotation angles
    return [aa_properties.get(aa, 0.0) * np.pi for aa in seq[:n_qubits]]
@qml.qnode(dev)
def protein_circuit(params, weights):
    for i, angle in enumerate(params):
        qml.RX(angle, wires=i)
    for i in range(n_qubits-1):
        qml.CNOT(wires=[i, i+1])
    for i in range(n_qubits):
        qml.RY(weights[i], wires=i)
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]
def protein_synthesis(seq):
    params = encode_sequence(seq)
    weights = np.random.uniform(0, np.pi, n_qubits)
    result = protein_circuit(params, weights)
    return {
        "Input Sequence": seq,
        "Encoded Angles": params,
        "Variational Weights": weights,
        "Protein Folding Signature": result
    }
user_input = input("Enter protein sequence (e.g., ACGK): ")
output = protein_synthesis(user_input)
print(output)
```
